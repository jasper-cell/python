# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成数据
x = np.arange(0,6,0.1)
y = np.sin(x)
```

```python
#绘制图形
plt.plot(x,y)
plt.show()
```

## pyplot功能

```python
#生成数据
x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#绘制图形
plt.plot(x,y1,label = 'sin')
plt.plot(x,y2,linestyle = '--',label = 'cos')#用虚线绘制
plt.xlabel("x")#x轴标签
plt.ylabel("y")#y轴标签
plt.title('sin & cos')
plt.legend()
plt.show()
```
